```python
import pandas as pd
from sklearn.utils import resample

# Load your dataset (replace with the actual path to your dataset)
data = pd.read_csv("creditcard.csv")

# Separate the dataset into legitimate and fraudulent transactions
legitimate_transactions = data[data['Class'] == 0]
fraudulent_transactions = data[data['Class'] == 1]

# Determine the size of the minority class (fraudulent transactions)
minority_class_size = len(fraudulent_transactions)

# Randomly undersample the majority class (legitimate transactions)
legitimate_sampled = resample(legitimate_transactions, n_samples=minority_class_size, random_state=42)

# Combine the sampled legitimate transactions with the fraudulent transactions
balanced_data = pd.concat([legitimate_sampled, fraudulent_transactions])

```
